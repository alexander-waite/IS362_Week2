# __Employment Rates Based on Education Level__ #

```python
import pandas as pd
```

#### Unemployment Data is from  __[Kaggle.com](https://www.kaggle.com/aniruddhasshirahatti/us-unemployment-dataset-2010-2020)__

```python
df = pd.read_csv("unemployment_data_us.csv")
```

```python
df.head(10)
```

#### The function below helps better organize the data by month.
It does this by using a dictionary of month to numeric value keys and then using the pandas apply function to rewrite values for each row in the Month column.

```python
def convert_month_to_num(month=str):
    month_num_key={'Jan':1,'Feb':2,'Mar':3,'Apr':4,'May':5,'Jun':6,'Jul':7,'Aug':8,'Sep':9,'Oct':10,'Nov':11,'Dec':12}
    month_str = month_num_key[month]
    return int(month_str)

df['Month'] = df['Month'].apply(convert_month_to_num)
```

The sort_values function will organize the columns in alpha-numeric values based on the header columns chosen

```python
df = df.sort_values(by=['Year','Month'],ascending=True)
df.head(13)
```

Finally I used pandas plot function to create a simple graph that shows the data in a line chart.

```python
df.plot(x='Year',y=['Primary_School','High_School','Associates_Degree','Professional_Degree'],figsize=(20,10),title='Unemployment percentage from 2010 to 2020 by education level')
```
